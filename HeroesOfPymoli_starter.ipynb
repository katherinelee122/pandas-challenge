## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,100]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40-100']

df['Age_Demographics'] = pd.cut(df["Age"], bins, labels=labels)
df = df.groupby("Age_Demographics")

total_count_age = df["Age"].nunique()
full_sum_players = df["Age"].sum()
age_demographics_df=pd.DataFrame(total_count_age)
percent_of_players = total_count_age / full_sum_players

age_demographics_df['Total Count Age'] = total_count_age
age_demographics_df2 = pd.DataFrame(age_demographics_df)
age_demographics_df2['Percent of Players'] = percent_of_players
age_demographics_df3 = pd.DataFrame(age_demographics_df2)
(age_demographics_df3)
# df = df.groupby("player_summary")
# total_players_value = df["SN"].count()
# total_count_player = df["SN"].sum()
# percentage_of_players = total_count_player/total_players_value

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9,14,19,24,29,34,39,100]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40-100']

df["Age_summary"] = pd.cut(df["Age"], bins, labels=labels)
df = df.groupby("Age_summary")
total_count_age = df["Purchase ID"].count()
age_purchase_df=pd.DataFrame(total_count_age)
avg_purchase_price_age = df ["Price"].mean()
total_purchase_value = df["Price"].sum()
total_count_age = df["SN"].nunique()
avg_purchase_per_person_age = total_purchase_value/total_count_age

age_purchase_df['Average Purchase Price'] = avg_purchase_price_age
age_purchase_df2 = pd.DataFrame(age_purchase_df)
age_purchase_df2['Total Purchase Value'] = total_purchase_value
age_purchase_df3 = pd.DataFrame(age_purchase_df2)
age_purchase_df3['Avg Total Purchase per Person'] = avg_purchase_per_person_age
age_purchase_df4 = pd.DataFrame(age_purchase_df3)
(age_purchase_df4)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
bins = [0,9,14,19,24,29,34,39,100]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40-100']

df["top_spenders"] = pd.cut(df["Price"],bins,labels=labels)

max_purchase_count = df['Price'].max()
avg_purchase_price = df['Price'].mean()
total_purchase_price = df['Price'].sum()
top_spenders_df=pd.DataFrame(max_purchase_count)

top_spenders_df['Max Purchase Price'] = max_purchase_price
top_spenders_df2 = pd.DataFrame(top_spenders_df)
top_spenders_df2['Average Purchase Price'] = avg_purchase_price
top_spenders_df3 = pd.DataFrame(top_spenders_df2)
top_spenders_df3['Total Purchase Price'] = total_purchase_price
top_spenders_df4 = pd.DataFrame(top_spenders_df3)
(top_spenders_df4)
```
